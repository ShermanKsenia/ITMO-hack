#### Ответ на открытый вопрос

```python
system_prompt = '''Ты — AI-ассистент для проверки открытых ответов студентов. Твоя задача — оценить, насколько ответ студента соответствует эталонному ответу, с учетом смысловой эквивалентности, а не дословного совпадения.

# Контекст
Ты получишь:
1. **Вопрос** — формулировка задания
2. **Эталонный ответ** — правильный, ожидаемый ответ
3. **Ответ студента** — ответ, который нужно проверить

# Особые случаи
- **Синонимия**: Засчитывай синонимичные формулировки
- **Частичная правильность**: Отмечай, какие части ответа верны
- **Дополнительная информация**: Если студент добавил верную дополнительную информацию — отметь это
- **Распространенные ошибки**: Выявляй типичные misconceptions
Ответь только "да", если ответ студента верный, и "нет" - если неверный'''
```

```python
json_input = {
    'question': "Что такое командная экономическая система?",
    'answer': "Экономическая система, при которой капитал и земля находятся в собственности государства, а распределение ресурсов осуществляется на основе директивных планов центральных органов управления.",
    'student_answer': 'В командной экономической системе капитал и земля находятся в собственности государства и используются директивные планы для распределения ресурсов',
}
```

```python
def get_user_prompt(json_input):
    user_question = f'''Вопрос: {json_input['question']}
    Правильный ответ: {json_input['answer']}
    Ответ студента: {json_input['student_answer']}'''
    return user_question

messages_right_answer = [
    {
        "role": 'system',
        'text': system_prompt
    },
    {
        'role': 'user',
        'text': get_user_prompt(json_input)
    }
]

sdk = YCloudML(
    folder_id=FOLDER_ID,
    auth=AUTH_TOKEN,
)

result = (
    sdk.models.completions("yandexgpt").configure(temperature=0.5).run(messages_right_answer)
)
```

```python
print(result.alternatives[0].text.strip('\n`'))
```
